```python
import networkx as nx
import json

with open("network_with_fake.json", "r") as f:
    data = json.load(f)
```

```python
playersOutperformance = {}
```

```python
def calculate_win_probability(rating_a, rating_b):
    return 1 / (1 + 10 ** ((rating_b - rating_a) / 400))
```

```python
games = data['links']
```

```python
for game in games:
    player1 = game['data'][0]['white']
    player1Username = player1['username']
    player1Rating = player1['rating']
    player1Result = player1['result']

    player2 = game['data'][0]['black']
    player2Username = player2['username']
    player2Rating = player2['rating']
    player2Result = player2['result']
    #get player score if it exists, otherwise 0
    if player1Username in playersOutperformance:
        player1Starting = playersOutperformance[player1Username]
    else:
        player1Starting = 0
    if player2Username in playersOutperformance:
        player2Starting = playersOutperformance[player2Username]
    else:
        player2Starting = 0

    player1WinProb = calculate_win_probability(player1Rating, player2Rating)
    player2WinProb = 1 - player1WinProb

    if (player1Result == 'win'):
        player1Realized = 1
    elif (player1Result == 'stalemate'):
        player1Realized = 0.5
    else:
        player1Realized = 0

    if (player2Result == 'win'):
        player2Realized = 1
    elif (player2Result == 'stalemate'):
        player2Realized = 0.5
    else:
        player2Realized = 0

    playersOutperformance[player1Username] = player1Starting + player1Realized - player1WinProb
    playersOutperformance[player2Username] = player2Starting + player2Realized - player2WinProb
```

```python
sortedOutperformance = {k: v for k, v in sorted(playersOutperformance.items(), key=lambda item: item[1], reverse=True)}
```
